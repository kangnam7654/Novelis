```python
from pathlib import Path
from datetime import datetime
import pandas as pd
```

# File Load

```python
dir = Path("./docs/")
fraud_path = dir.joinpath("fraud.csv")
ip_path = dir.joinpath("ip_country.xlsx")

#
fraud_file = pd.read_csv(fraud_path, index_col=0)
ip_file = pd.read_excel(ip_path)

#
fraud_file.head()
```

```python
def to_datetime(x):
    time_format = "%m/%d/%Y %H:%M"
    return datetime.strptime(x, time_format)
```

```python
fraud_file["signup_time"] = fraud_file["signup_time"].apply(to_datetime)
fraud_file["purchase_time"] = fraud_file["purchase_time"].apply(to_datetime)
```

```python
fraud_file
```

```python
fraud_file.insert(loc=3, column="time_delta", value=None)
```

```python
fraud_file["time_delta"] = fraud_file["purchase_time"] - fraud_file["signup_time"]
```

```python
fraud_file["time_delta"].iloc[0].total_seconds() / 60
```

```python
def to_minuets(x):
    return x.total_seconds() / 60
```

```python
fraud_file["time_delta"] = fraud_file["time_delta"].apply(to_minuets)
```
